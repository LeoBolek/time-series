# Analisando assinantes da newsletter da Empresa X

## Importando as bibliotecas

```python
!pip install seaborn
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importando os dados

```python
assinantes = pd.read_csv('dados/newsletter_empresa_x.csv')
assinantes
```

```python
# Ajustando os nomes das colunas por questão de estética

ajuste_col = {
    'mes' : 'Mês',
    'assinantes' : 'Assinantes'
}

assinantes = assinantes.rename(columns = ajuste_col)
assinantes
```

```python
assinantes.dtypes
```

```python
print(f'Quantidade de linhas e colunas: {assinantes.shape}')
print(f'Quantidade de dados nulos: {assinantes.isna().sum().sum()}')
```

```python
# Convertendo o 'Mês' para datetime

assinantes['Mês'] = pd.to_datetime(assinantes['Mês'])
assinantes.dtypes
```

## Analisando o aumento de assinantes

```python
assinantes['Aumento'] = assinantes['Assinantes'].diff()
assinantes
```

## Analisando a aceleração de assinantes

```python
assinantes['Aceleração'] = assinantes['Aumento'].diff()
assinantes
```

## Comparando o número, aumento e aceleração de assinantes

```python
def grupo_plot(x, y1, y2, y3, dataset, titulo):
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    plt.figure(figsize = (16, 12))
    ax = plt.subplot(3, 1, 1)
    ax.set_title(titulo, fontsize = 18, x = 0.40, y = 1)
    sns.lineplot(x = x, y = y1, data = dataset)
    plt.subplot(3, 1, 2)
    sns.lineplot(x = x, y = y2, data = dataset)
    plt.subplot(3, 1, 3)
    sns.lineplot(x = x, y = y3, data = dataset)
    ax = ax
```

```python
grupo_plot('Mês', 'Assinantes', 'Aumento', 'Aceleração', assinantes,
           'Análise comparativa de número, aumento e aceleração de assinantes, respectivamente')
```

**Podemos notar que o número de assinantes obteve um crescimento expressivo. 
Já no aumento de assinantes, entre o mês de Janeiro e Abril de 2018 estagnou, e após Abril de 2018 o aumento de assinantes caiu.
Na aceleração houve um crescimento inicial, entretanto, a partir de Outubro de 2017 a aceleração ficou negativa e caiu muito, recuperando-se e obtendo um aumento expressivo apenas após Outubro de 2018.**
