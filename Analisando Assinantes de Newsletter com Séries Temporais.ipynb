## Importando os dados

```python
assinantes = pd.read_csv('dados/newsletter_empresa_x.csv')
assinantes
```

```python
# Ajustando os nomes das colunas por questÃ£o de estÃ©tica

ajuste_col = {
    'mes' : 'MÃªs',
    'assinantes' : 'Assinantes'
}

assinantes = assinantes.rename(columns = ajuste_col)
assinantes
```

```python
assinantes.dtypes
```

```python
print(f'Quantidade de linhas e colunas: {assinantes.shape}')
print(f'Quantidade de dados nulos: {assinantes.isna().sum().sum()}')
```

```python
# Convertendo o 'MÃªs' para datetime

assinantes['MÃªs'] = pd.to_datetime(assinantes['MÃªs'])
assinantes.dtypes
```

## Analisando o aumento de assinantes

```python
assinantes['Aumento'] = assinantes['Assinantes'].diff()
assinantes
```

## Analisando a aceleraÃ§Ã£o de assinantes

```python
assinantes['AceleraÃ§Ã£o'] = assinantes['Aumento'].diff()
assinantes
```

## Comparando o nÃºmero, aumento e aceleraÃ§Ã£o de assinantes

```python
def grupo_plot(x, y1, y2, y3, dataset, titulo):
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    plt.figure(figsize = (16, 12))
    ax = plt.subplot(3, 1, 1)
    ax.set_title(titulo, fontsize = 18, x = 0.40, y = 1)
    sns.lineplot(x = x, y = y1, data = dataset)
    plt.subplot(3, 1, 2)
    sns.lineplot(x = x, y = y2, data = dataset)
    plt.subplot(3, 1, 3)
    sns.lineplot(x = x, y = y3, data = dataset)
    ax = ax
```

```python
grupo_plot('MÃªs', 'Assinantes', 'Aumento', 'AceleraÃ§Ã£o', assinantes,
           'AnÃ¡lise comparativa de nÃºmero, aumento e aceleraÃ§Ã£o de assinantes, respectivamente')
```

**Podemos notar que o nÃºmero de assinantes obteve um crescimento expressivo. 
JÃ¡ no aumento de assinantes, entre o mÃªs de Janeiro e Abril de 2018 estagnou, e apÃ³s Abril de 2018 o aumento de assinantes caiu.
Na aceleraÃ§Ã£o houve um crescimento inicial, entretanto, a partir de Outubro de 2017 a aceleraÃ§Ã£o ficou negativa e caiu muito, recuperando-se e obtendo um aumento expressivo apenas apÃ³s Outubro de 2018.**
