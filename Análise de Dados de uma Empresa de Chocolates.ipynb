## Importando os dados

```python
chocolar = pd.read_csv('dados/chocolar.csv')
chocolar
```

```python
nomes_col = {
    'mes' : 'Mês',
    'vendas' : 'Vendas'
}
```

```python
chocolar = chocolar.rename(columns = nomes_col)
chocolar
```

```python
chocolar.dtypes
```

```python
chocolar['Mês'] = pd.to_datetime(chocolar['Mês'])
chocolar.dtypes
```

```python
print(f'Quantidade de linhas e colunas: {chocolar.shape}')
print(f'Quantidade de dados nulos: {chocolar.isna().sum().sum()}')
```

```python
chocolar['Aumento'] = chocolar['Vendas'].diff()
chocolar['Aceleração'] = chocolar['Aumento'].diff()
chocolar
```

```python
 def plot_comparacao(x, y1, y2, y3, dataset, titulo):
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    plt.figure(figsize = (16, 12))
    ax = plt.subplot(3, 1, 1)
    ax.set_title(titulo, fontsize = 16, x = 0.27, y = 1)
    sns.lineplot(x = x, y = y1, data = dataset)
    plt.subplot(3, 1, 2)
    sns.lineplot(x = x, y = y2, data = dataset)
    plt.subplot(3, 1, 3)
    sns.lineplot(x = x, y = y3, data = dataset)
    ax = ax
```

```python
plot_comparacao('Mês', 'Vendas', 'Aumento', 'Aceleração', 
                chocolar, 'Análise de vendas da Chocolar de 2017 a 2018')
```

**Podemos notar uma sazonalidade nos três gráficos, principalmente nos períodos de Abril de 2017 a Junho de 2017 e Abril de 2018 a Junho de 2018. É importante salientar que Abril é o mês da Páscoa e Junho tem o Dia dos Namorados.**

# Vendas diárias da Chocolar (Outubro e Novembro)

```python
vendas_diarias = pd.read_csv('dados/vendas_por_dia.csv')
vendas_diarias
```

```python
nomes_col2 = {
    'dia' : 'Dia',
    'vendas' : 'Vendas'
}

vendas_diarias = vendas_diarias.rename(columns = nomes_col2)
vendas_diarias.head()
```

```python
print(f'Quantidade de linhas e colunas: {vendas_diarias.shape}')
print(f'Quantidade de dados nulos: {vendas_diarias.isna().sum().sum()}')
```

```python
vendas_diarias.dtypes
```

```python
vendas_diarias['Dia'] = pd.to_datetime(vendas_diarias['Dia'])
vendas_diarias.dtypes
```

```python
vendas_diarias['Aumento'] = vendas_diarias['Vendas'].diff()
vendas_diarias['Aceleração'] = vendas_diarias['Aumento'].diff()
vendas_diarias
```

```python
def plot_vendas(x, y1, y2, y3, dataset, titulo):
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    plt.figure(figsize = (16, 12))
    ax = plt.subplot(3, 1, 1)
    ax.set_title(titulo, fontsize = 16, x = 0.27, y = 1)
    sns.lineplot(x = x, y = y1, data = dataset)
    plt.subplot(3, 1, 2)
    sns.lineplot(x = x, y = y2, data = dataset)
    plt.subplot(3, 1, 3)
    sns.lineplot(x = x, y = y3, data = dataset)
    ax = ax
    
plot_vendas('Dia', 'Vendas', 'Aumento', 'Aceleração',
           vendas_diarias, 'Análise de vendas da Chocolar de Outubro a Novembro de 2018')
```

## Analisando a sazonalidade

```python
# Adicionando dias da semana

vendas_diarias['Dia_da_semana'] = vendas_diarias['Dia'].dt.day_name()
vendas_diarias
```

```python
vendas_diarias['Dia_da_semana'].unique()
```

```python
# Traduzindo os dias da semana

traducao_dias = {
    'Monday' : 'Segunda',
    'Tuesday' : 'Terça',
    'Wednesday' : 'Quarta',
    'Thursday' : 'Quinta',
    'Friday' : 'Sexta',
    'Saturday' : 'Sábado',
    'Sunday' : 'Domingo'
}
```

```python
vendas_diarias['Dia_da_semana'] = vendas_diarias['Dia_da_semana'].map(traducao_dias)
vendas_diarias
```

## Agrupando os dias

```python
# Pegando a média das vendas, aumento e aceleração dos dias da semana e arredondando os valores

vendas_diarias_agrupadas = vendas_diarias.groupby('Dia_da_semana')[['Vendas', 'Aumento', 'Aceleração']].mean().round()
vendas_diarias_agrupadas
```

**As quedas da sazonalidade das vendas são referentes aos finais de semana. Talvez pela empresa trabalhar meio turno ou porque tem menos vendedores. Também podemos levar em conta o fato de que mais pessoas vão para a praia ou visitar parentes em outras cidades nos finais de semana.**

## Correlação das venhas diárias

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlação das vendas diárias', fontsize=18, x=0.3, y=0.95)
autocorrelation_plot(vendas_diarias['Vendas'])
ax = ax
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlação do aumento das vendas diárias', fontsize = 18,  x= 0.3, y = 0.95)
autocorrelation_plot(vendas_diarias['Aumento'][1:])
ax = ax
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlação da aceleração das vendas diárias', fontsize = 18,  x= 0.3, y = 0.95)
autocorrelation_plot(vendas_diarias['Aceleração'][2:])
ax = ax
```

**Existe uma correlação das vendas diárias, do aumento das vendas diárias e da aceleração das vendas diárias.**

# Statsmodels

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
resultado = seasonal_decompose(chocolar['Vendas'], period = 3)
ax = resultado.plot()
```

```python
observacao = resultado.observed
tendencia = resultado.trend
sazonalidade = resultado.seasonal
ruido = resultado.resid
```

```python
data = ({
    'Observação' : observacao,
    'Tendência' : tendencia,
    'Sazonalidade' : sazonalidade,
    'Ruído' : ruido
})

resultado = pd.DataFrame(data)
resultado
```

```python
plot_comparacao(resultado.index, 'Observação', 'Tendência', 'Sazonalidade',
               resultado, 'Exemplo de Statsmodels')
```
